```python
# 문자열 뿐만 아니라 숫자인 값들도 데이터 타입이 object(string)이기 때문에 파일로 저장 후 다시 불러옴

weather.info()
```

```python
# 숫자 데이터가 int와 float으로 적절하게 변환 됨

weather.to_csv('data/weather_2023.csv', index=False)
weather = pd.read_csv('data/weather_2023.csv')
weather.info()
```

```python
# 모든 데이터의 컬럼을 확인하고

weather.columns.sort_values()
```

```python
# NaN 값을 0으로 채움

weather['sumRn'] = weather['sumRn'].fillna(0)
weather
```

```python
# 날짜 데이터 중 월, 일을 사용하기 편리하도록 분리해줄 예정
# tm은 object(string)으로 저장되어 있음

print(weather["tm"].info())
```

```python
# object(string) 형식에서 datetime(날짜) 형식으로 변환

weather['tm'] = pd.to_datetime(weather['tm'])
print(weather["tm"].info())
```

```python
# 년, 월, 일을 분리하여 각각 year, month, day라는 컬럼을 뒤에 만들어서 삽입

weather.loc[:,'year'] = weather['tm'].dt.year
weather.loc[:,'month'] = weather['tm'].dt.month
weather.loc[:,'day'] = weather['tm'].dt.day
weather
```

```python
# 필요한 데이터 컬럼을 뽑음
# tm = 날짜, avgTa = 일 평균 기온, sumRn = 일 강수량, avgWs = 일 평균 풍속

weather = weather[['avgTa', 'sumRn', 'avgWs', 'year', 'month', 'day']]
weather
```

```python
# 전처리 완료한 날씨 데이터프레임 저장
weather.to_csv('data/weather.csv', index=False)
```

# 따릉이 대여소 데이터 전처리

```python
# 데이터 출처 : 서울 열린데이터 광장
# 서울시 공공자전거 대여소 정보 : http://data.seoul.go.kr/dataList/OA-13252/F/1/datasetView.do

rental = pd.read_csv('data/originalRental.csv', encoding='cp949')
rental
```

```python
# 필요한 데이터 만 추출

rental = rental[['대여소\n번호', '보관소(대여소)명', '소재지(위치)', 'Unnamed: 4', 'Unnamed: 5']]
rental
```

```python
# 컬럼 명 정리

rental = rental.rename(columns={'대여소\n번호' : '대여소번호', '보관소(대여소)명' : '대여소명', '소재지(위치)' : "자치구", "Unnamed: 3" : "상세주소", "Unnamed: 4" : "위도", "Unnamed: 5" : "경도"})
rental
```

```python
# 필요없는 행 삭제

rental = rental.drop([0, 1, 2, 3])
rental
```

```python
# 인덱스 재정의

rental = rental.reset_index()
rental
```

```python
# 기존 인덱스 컬럼 삭제

rental = rental.drop(["index"], axis=1)
rental
```

```python
# 대여소변호 컬럼 데이터의 float형을 int형으로 변환

rental['대여소번호'] = rental['대여소번호'].astype(int)
rental
```

```python
# 전처리 완료한 따릉이 대여소 데이터 저장

rental.to_csv('data/rental.csv', index=False)
```

```python
# 대여소번호를 key, 자치구를 value로 갖는 딕셔너리 생성
number_local_dict = rental.set_index('대여소번호')['자치구'].to_dict()

# 생성한 딕셔너리 확인
number_local_dict
```

# 따릉이 이용 데이터 전처리  
#  


```python
# 데이터 출처 : 서울 열린데이터 광장
# 서울시 공공자전거 이용정보(일별) : http://data.seoul.go.kr/dataList/OA-15246/F/1/datasetView.do

# 웗 별로분리되어 있는 파일들 읽어오기


bicycle1 = pd.read_csv('data/bicycle1.csv', encoding='cp949')
bicycle2 = pd.read_csv('data/bicycle2.csv', encoding='cp949')
bicycle3 = pd.read_csv('data/bicycle3.csv', encoding='cp949')
bicycle4 = pd.read_csv('data/bicycle4.csv', encoding='cp949')
bicycle5 = pd.read_csv('data/bicycle5.csv', encoding='cp949')
bicycle6 = pd.read_csv('data/bicycle6.csv', encoding='cp949')
```

```python
# concat으로 1 ~ 6월 이용 내역 병합

bicycle = pd.concat([bicycle1, bicycle2, bicycle3, bicycle4, bicycle5, bicycle6])
bicycle
```

```python
# 인덱스 재정의 (데이터들의 중복 인덱스 제거)

bicycle = bicycle.reset_index()
bicycle
```

```python
# 잘못된 기존 인덱스 컬럼 삭제

bicycle = bicycle.drop(["index"], axis=1)
bicycle
```

```python
# object(string) 형태인 대여일자 열을 datetime 형태로 변환하고 년, 월, 일을 분리하여 각각 year, month, day라는 컬럼을 뒤에 만들어서 삽입

bicycle['대여일자'] = pd.to_datetime(bicycle['대여일자'])
bicycle.loc[:,'year'] = bicycle['대여일자'].dt.year
bicycle.loc[:,'month'] = bicycle['대여일자'].dt.month
bicycle.loc[:,'day'] = bicycle['대여일자'].dt.day
bicycle
```

```python
# 필요한 열 만 추출

bicycle = bicycle[['대여소번호', '대여소', '성별', '연령대', '운동량', 'year', 'month', 'day']]
bicycle
```

```python
# 대여소의 설치 위치를 데이터프레임에 추가하기 위해 비어있는 컬럼 미리 추가

bicycle['자치구'] = ""
bicycle
```

```python
# 대여소 데이터 파싱 때 만들어 둔 딕셔너리를 이용하여 행 마다 대여소번호를 key로 자치구 value값을 찾아서 자치구 열에 삽입
# 없어진 대여소는 '철거된 대여소' 삽입


a = bicycle.loc[:, '대여소번호'] # 대여소 번호 열 리스트 생성

for i, item in enumerate(a): # i = 인덱스, item = 해당 인덱스의 대여소 번호
    bicycle.loc[i, '자치구'] = number_local_dict.get(item, "철거된 대여소")

bicycle
```

```python
data_type_of_exercise = bicycle['운동량'].dtype
print(f"'운동량' 열의 데이터 타입: {data_type_of_exercise}")
```

```python
bicycle['운동량'] = pd.to_numeric(bicycle['운동량'], errors='coerce')

data_type_of_exercise = bicycle['운동량'].dtype
print(f"'운동량' 열의 데이터 타입: {data_type_of_exercise}")
```

```python
# 전처리 완료한 따릉이 이용 데이터 저장

bicycle.to_csv('data/bicycle.csv', index=False)
```

# 체력 측정 데이터 전처리

```python
# 데이터 출처 : 국가 통계포털
# 연령별 체력, 체격 측정결과 : 서울(남자) 2017년 : https://kosis.kr/statHtml/statHtml.do?orgId=113&tblId=DT_113_STBL_1025785&conn_path=I2
# 연령별 체력, 체격 측정결과 : 서울(여자) 2017년 : https://kosis.kr/statHtml/statHtml.do?orgId=113&tblId=DT_113_STBL_1025783&conn_path=I2

male = pd.read_csv('data/maleStrength.csv', encoding='utf-8')
male
female = pd.read_csv('data/femaleStrength.csv', encoding='utf-8')
female
```

```python
# 첫 번째 행을 컬럼 명으로 교체, 첫 번째 행 삭제

male.columns = male.iloc[0]
male = male[1:]
male
female.columns = female.iloc[0]
female = female[1:]
female
```

```python
# 70대부터의 데이터는 없지만 체력평가 중 자전거와 가장 비슷한 20m왕복오래달리기(회) 평균 데이터 만 추출

male = male = pd.DataFrame([male.loc[10]])
male
female = pd.DataFrame([female.loc[10]])
female
```

```python
# concat으로 남성과 여성 데이터 결합
strength =  pd.concat([male, female])
strength
```

```python
# 필요한 데이터의 컬럼 만 추출

strength = strength.iloc[:, :11]
strength
```

```python
# 성별에 맞게 인덱스 변경

strength.index = [0, 1]
strength
```

```python
strength.insert(0, '성별', ['남성', '여성'])
strength
```

```python
# 전처리 완료한 체력 데이터 저장

strength.to_csv('data/strength.csv', index=False)
```
