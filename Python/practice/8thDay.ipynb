```python
import folium
```

```python
m = folium.Map(location=[37.54392611998678, 126.95105440539045], zoom_start=16)
# CircleMarker : 상대적 원
folium.CircleMarker([37.54392611998678, 126.95105440539045], radius=100, color='#34EBC6', fill_color='#C634EB', popup='CircleMarker', tooltip='tooltip').add_to(m)
# Circle : 절대적 원
folium.Circle([37.54392611998678, 126.95105440539045], radius=100, color='#EB9E34', fill_color='red', popup='Circle', tooltip='tooltip').add_to(m)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print(df)

# Iterating through rows using iterrows()
for index, row in df.iterrows():
    print(f"Index: {index}")
    print(f"Name: {row['Name']}, Age: {row['Age']}, City: {row['City']}")
    print("-" * 30)
```

```python
# https://fontawesome.com/icons
taxiData = pd.read_csv("택시승차대 현황.csv", encoding='cp949')
taximap = folium.Map(location=[taxiData['위도'][0],taxiData['경도'][0]], zoom_start=15)
for i in range(10):
    folium.Marker(location=[taxiData['위도'][i],taxiData['경도'][i]]
                  , popup = folium.Popup(taxiData["위치명"][i], max_width=650)
                  , tooltip = taxiData["위치명"][i]
                  , icon=folium.Icon(color = "blue", icon = 'fa-solid fa-taxi', prefix = 'fa-solid')).add_to(taximap)
    folium.Circle([taxiData['위도'][i],taxiData['경도'][i]], radius=200, color='#EB9E34', fill_color='red', popup='Circle', tooltip='tooltip').add_to(taximap)

taximap
```
