```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
plt.plot([0, 1, 2, 3], [1, 2, 3, 4])
plt.show()
```

```python
plt.plot([1, 2, 3, 4, 6])
plt.axis([-1, 5, 0, 7])

plt.title("", loc="center", pad = 20)

plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, cmap="Spectral", alpha=0.5)
plt.show()
```

```python
ratio = [34,32,16,18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
wedgeprops = { 'width' : 0.7, 'edgecolor':'k', 'linewidth':1}

plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=[0, 0.1, 0, 0.1], colors=['red', 'yellow', 'green', 'purple'], wedgeprops=wedgeprops)
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.figure(figsize=(16,9))
plt.plot(months, sales_2019, label = "2019", ls='--', c='C9', marker='o')
plt.plot(months, sales_2020, label = "2020", ls='-.', c='C4', marker='^')
plt.grid(alpha=0.5)
plt.xlabel("")
plt.ylabel("")

plt.legend()
plt.title("   (2019-2020)")
plt.show()
```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]


plt.figure(figsize=(16,9))
plt.plot(categories, data, dash_capstyle="round", c='C8', marker='X', label = " ")
plt.bar(categories, data, width= 0.5, color = ['C0','C1','C2','C3','C4'], label = " ")
plt.ylim(0,50)
plt.grid(alpha = 0.5)
plt.legend()
plt.xlabel("", size = 20)
plt.xticks(rotation=45)
plt.ylabel("", size = 20, rotation = 0)
plt.title(" ", size = 30)

plt.show()
```

```python
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # Banana   
plt.figure(figsize=(16,9))
plt.pie(sizes, explode = explode, labels=labels, autopct="%.1f%%", startangle = 90, shadow = True)
plt.legend()
plt.summer()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize=(16,9))
plt.plot(x, sin, label = "sin(x)", c = "lightcoral")
plt.plot(x, cos, label = "cos(x)", c = "slateblue")
plt.legend(fontsize = 15)
plt.xlabel("X")
plt.ylabel("Y", rotation = 0)
plt.grid(alpha = 0.5)
plt.title(",  ")

plt.show()
```
